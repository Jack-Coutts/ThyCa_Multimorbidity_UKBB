```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import re
import seaborn as sns

```

```python
# Globally set the graph style and font size

sns.set_style('darkgrid') # darkgrid, white grid, dark, white and ticks
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=15)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=16)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
plt.rc('legend', fontsize=14)    # legend fontsize
plt.rc('font', size=14)          # controls default text sizes
```

```python
# Read in the tsv's as dataframes

levo = pd.read_csv('levo_FA_dfg.tsv', sep='\t', index_col=0)

th_prod = pd.read_csv('tp_FA_dfg.tsv', sep='\t', index_col=0)

th_sod = pd.read_csv('ts_FA_dfg.tsv', sep='\t', index_col=0)

th_based = pd.read_csv('thyrox_based_FA_dfg.tsv', sep='\t', index_col=0)
```

```python
# Plot Figure

fig, axes = plt.subplots(2, 2, figsize=(25, 20), tight_layout=True)

#fig.suptitle('Thyroxine-based medications')

# Create colour palette
palette1 = sns.color_palette('autumn', 70) #len(psar_ThyCa1_5['lift'])) = 35. Using 70 = larger range of colours so can choose middle shades
#palette.reverse()

b1 = sns.barplot(ax=axes[0,1], x=levo['lift'],
                 y=levo['Antecedent Disease'],
                 palette=np.array(palette1[29:70]), # Only selecting the middle shades - the extremes are too light or dark
                 ci=None)
# Add bar values
b1.bar_label(b1.containers[0])

# Set axis labels and title
b1.set(title='Levothyroxine Disease Associations', xlabel='Lift', ylabel='Disease Name')


# Create colour palette
palette2 = sns.color_palette('Reds_r', 70) #len(psar_ThyCa1_5['lift'])) = 35. Using 70 = larger range of colours so can choose middle shades
#palette.reverse()

b2 = sns.barplot(ax=axes[1,0], x=th_prod['lift'],
                 y=th_prod['Antecedent Disease'],
                 palette=np.array(palette2[40:63]), # Only selecting the middle shades - the extremes are too light or dark
                 ci=None)
# Add bar values
b2.bar_label(b2.containers[0])

# Set axis labels and title
b2.set(title='Thyroxine Product Disease Associations', xlabel='Lift', ylabel='Disease Name')


# Create colour palette
palette3 = sns.color_palette('bone', 70) #len(psar_ThyCa1_5['lift'])) = 35. Using 70 = larger range of colours so can choose middle shades
#palette.reverse()

b3 = sns.barplot(ax=axes[1,1], x=th_sod['lift'],
                 y=th_sod['Antecedent Disease'],
                 palette=np.array(palette3[55:70]), # Only selecting the middle shades - the extremes are too light or dark
                 ci=None)
# Add bar values
b3.bar_label(b3.containers[0])

# Set axis labels and title
b3.set(title='Thyroxine Sodium Disease Associations', xlabel='Lift', ylabel='Disease Name')


# Create colour palette
palette4 = sns.color_palette('Blues_r', 70) #len(psar_ThyCa1_5['lift'])) = 35. Using 70 = larger range of colours so can choose middle shades
#palette.reverse()

b4 = sns.barplot(ax=axes[0,0], x=th_based['lift'],
                 y=th_based['Antecedent Disease'],
                 palette=np.array(palette4[15:70]), # Only selecting the middle shades - the extremes are too light or dark
                 ci=None)
# Add bar values
b4.bar_label(b4.containers[0])

# Set axis labels and title
b4.set(title='Thyroxine-based Medication Disease Associations', xlabel='Lift', ylabel='Disease Name')


fig.savefig("sep_thyrox_meds.png") 
```
